# Top 10 Neighborhoods With the Least Arrest Rates

```python

import pandas as pd

DATA_PATH = "/Users/amitvapal/Downloads/pbp_arrests_2024_jan1_thru_nov30_2025.xlsx"


```

```python
df = pd.read_excel(DATA_PATH)
print("Loaded shape:", df.shape)

```

```python
df["Neighborhood"] = df["Neighborhood"].astype("string").fillna("Unknown")

neighborhood_counts = df["Neighborhood"].value_counts().sort_values(ascending=True)

```

```python
least_10 = neighborhood_counts.head(10)
least_10
```

```python
import matplotlib.pyplot as plt

# Plot the top 10 neighborhoods with the least arrests
plt.figure(figsize=(10, 6))
least_10.sort_values().plot(
    kind="barh",
    color="skyblue",
    edgecolor="black"
)

plt.title("Top 10 Neighborhoods With the Least Arrest Rates", fontsize=14)
plt.xlabel("Number of Arrests", fontsize=12)
plt.ylabel("Neighborhood", fontsize=12)
plt.tight_layout()
plt.show()
```
